```python
# import openpyxl and tkinter modules
from openpyxl import *
from tkinter import *
from tkinter import messagebox
 
# globally declare wb and sheet variable
 
# opening the existing excel file
wb = load_workbook('./donnees.xlsx')
 
# create the sheet object
sheet = wb.active
 
 
def excel():
     
    # resize the width of columns in
    # excel spreadsheet
    sheet.column_dimensions['A'].width = 20
    sheet.column_dimensions['B'].width = 20
    sheet.column_dimensions['C'].width = 20
    sheet.column_dimensions['D'].width = 20
    sheet.column_dimensions['E'].width = 20
 
    # write given data to an excel spreadsheet
    # at particular location
    
    sheet.cell(row=1, column=1).value = "élément"
    sheet.cell(row=1, column=2).value = "donnée1"
    sheet.cell(row=1, column=3).value = "donnée2"
    sheet.cell(row=1, column=4).value = "itérations"
    sheet.cell(row=1, column=5).value = "résultat"

# Function to set focus
def focus1(event):
    # set focus on the form_no_field box
    name_field.focus_set()
 
# Function to set focus (cursor)
def focus2(event):
    # set focus on the course_field box
    donnee1_field.focus_set()
 
 
# Function to set focus
def focus3(event):
    # set focus on the sem_field box
    donnee2_field.focus_set()
 
 
# Function to set focus
def focus4(event):
    # set focus on the form_no_field box
    iteration_field.focus_set()
 
# Function to set focus
def focus5(event):
    # set focus on the form_no_field box
    resultat_field.focus_set()
 
 
# Function for clearing the
# contents of text entry boxes
def clear():
     
    # clear the content of text entry box
    name_field.delete(0, END)
    donnee1_field.delete(0, END)
    donnee2_field.delete(0, END)
    iteration_field.delete(0, END)
    resultat_field.delete(0, END)
 
 
# Function to take data from GUI
# window and write to an excel file
def insert():
     
    # if user not fill any entry
    # then print "empty input"
    if (name_field.get() == "" and
        donnee1_field.get() == "" and
        donnee2_field.get() == "" and
        iteration_field.get() == "" and
        resultat_field.get() == "" ):

             
        print("empty input")
 
    else:
 
        # assigning the max row and max column
        # value upto which data is written
        # in an excel sheet to the variable
        current_row = sheet.max_row
        current_column = sheet.max_column
 
        # get method returns current text
        # as string which we write into
        # excel spreadsheet at particular location
        sheet.cell(row=current_row + 1, column=1).value = name_field.get()
        sheet.cell(row=current_row + 1, column=2).value = donnee1_field.get()
        sheet.cell(row=current_row + 1, column=3).value = donnee2_field.get()
        sheet.cell(row=current_row + 1, column=4).value = iteration_field.get()
        sheet.cell(row=current_row + 1, column=5).value = resultat_field.get()
 
        # save the file
        wb.save('./donnees.xlsx')
 
        # set focus on the name_field box
        name_field.focus_set()
 
        # call the clear() function
        clear()
        
    
 
# Driver code
if __name__ == "__main__":
     
    # create a GUI window
    root = Tk()
 
    # set the background colour of GUI window
    root.configure(background='#41B77F')
    
 
    # set the title of GUI window
    root.title("Palantir solution")
 
    # set the configuration of GUI window
    root.geometry("500x300")
    
    def msg():
        messagebox.showwarning("Palantir solution","Les paramètres ont été rentrés avec succès")

 
    excel()
 
    # create a Form label
    heading = Label(root, text="ENTREZ VOS DONNEES FLOTTANTES", bg="#41B77F")
 
    # create a Name label
    name = Label(root, text="Name", bg="#41B77F")
 
    # create a donnee1 label
    donnee1 = Label(root, text="donnée1", bg="#41B77F")
 
    # create a donnee2 label
    donnee2 = Label(root, text="donnée2", bg="#41B77F")
 
    # create a iteration label
    iteration = Label(root, text="Itération", bg="#41B77F")
    
    # create a resultat label
    resultat = Label(root, text="résultat", bg="#41B77F")
 
    # grid method is used for placing
    # the widgets at respective positions
    # in table like structure .
    heading.grid(row=0, column=1)
    name.grid(row=1, column=0)
    donnee1.grid(row=2, column=0)
    donnee2.grid(row=3, column=0)
    iteration.grid(row=4, column=0)
    resultat.grid(row=5, column=0)
 
    # create a text entry box
    # for typing the information
    name_field = Entry(root)
    donnee1_field = Entry(root)
    donnee2_field = Entry(root)
    iteration_field = Entry(root)
    resultat_field = Entry(root)
 
    # bind method of widget is used for
    # the binding the function with the events
 
    # whenever the enter key is pressed
    # then call the focus1 function
    name_field.bind("<Return>", focus1)
 
    # whenever the enter key is pressed
    # then call the focus2 function
    donnee1_field.bind("<Return>", focus2)
 
    # whenever the enter key is pressed
    # then call the focus3 function
    donnee2_field.bind("<Return>", focus3)
 
    # whenever the enter key is pressed
    # then call the focus4 function
    iteration_field.bind("<Return>", focus4)
    
    # whenever the enter key is pressed
    # then call the focus4 function
    resultat_field.bind("<Return>", focus5)
 
    # grid method is used for placing
    # the widgets at respective positions
    # in table like structure .
    name_field.grid(row=1, column=1, ipadx="100")
    donnee1_field.grid(row=2, column=1, ipadx="100")
    donnee2_field.grid(row=3, column=1, ipadx="100")
    iteration_field.grid(row=4, column=1, ipadx="100")
    resultat_field.grid(row=5, column=1, ipadx="100")
 
    # call excel function
    excel()
 
    # create a Submit Button and place into the root window
    submit = Button(root, text="Submit", fg="White", font=("Courrier", 25),
                            bg="Black", command= lambda:[insert(), msg()])
    submit.grid(row=8, column=1)
 
 

    
   
    # start the GUI
    root.mainloop()
    
```
